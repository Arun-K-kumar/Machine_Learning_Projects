```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
import pandas as pd
import torch 
import torch.nn as nn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
device = 'cuda' if torch.cuda.is_available else 'cpu' #not necessary for this model thus not used here
```

```python
f = pd.read_csv("Cancer_Data.csv") #Dataset URL -> https://www.kaggle.com/datasets/erdemtaha/cancer-data
f.head()
```

```python
f = f.drop("id",axis=1)
```

```python
f = f.drop("Unnamed: 32",axis=1)
```

```python
out = f["diagnosis"]
f = f.drop("diagnosis",axis=1)
```

```python
label = []
for i in range(len(out)):
    if out[i] == 'M':
        label.append(0)
    else:
        label.append(1)
```

```python
label[0]
```

```python
scaler = StandardScaler()
f = scaler.fit_transform(f)
```

```python
X=f
Y = label
```

```python
X.shape, Y[:5]
```

```python
X_train,X_test,Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state=42)
```

```python
X_train = torch.tensor(X_train, dtype = torch.float)
Y_train = torch.tensor(Y_train, dtype = torch.float).unsqueeze(1)
X_test = torch.tensor(X_test, dtype=torch.float)
Y_test = torch.tensor(Y_test, dtype=torch.float).unsqueeze(1)
```

```python
class LinearRegression(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1 = nn.Linear(in_features=30,out_features=1)
        

    def forward(self, x):
        return self.layer_1(x)
```

```python
torch.manual_seed(10)
model = LinearRegression()
loss_func = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(params=model.parameters(), lr =0.01)
```

```python
model.state_dict()
```

```python

epochs =  5000
model.load_state_dict(torch.load("Cancer_pred_model.pth"))
for epoch in range(epochs):
    model.train()
    Y_logits = model(X_train)
    loss = loss_func(Y_logits, Y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    model.eval()
    with torch.inference_mode():
        Y_test_logits = model(X_test)
        test_loss = loss_func(Y_test_logits, Y_test)

    if epoch%1000 == 0:
        print(f"Epoch {epoch} | Train Loss {loss} | Test Loss :{test_loss}")

torch.save(model.state_dict(), "Cancer_pred_model.pth")
```
